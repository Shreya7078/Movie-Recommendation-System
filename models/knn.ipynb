```python
from difflib import get_close_matches

def find_movie_name(movie_name, df):
    movie_name = movie_name.lower()
   
    titles = df['title'].str.lower().tolist()

    if movie_name in titles:
        return movie_name  

    partial = df[df['title'].str.lower().str.contains(movie_name)]
    if not partial.empty:
        return partial['title'].iloc[0].lower()

    match = get_close_matches(movie_name, titles, n=1, cutoff=0.4)
    if match:
        return match[0]

    return None

```

```python
def recommend(movie_name, df, tfidf_matrix, knn):
    
    corrected_name = find_movie_name(movie_name, df)

    if corrected_name is None:
        print("❌ Movie not found. Please check spelling or try another movie.")
        return
    
    movie_name = corrected_name

    idx = df[df['title'].str.lower() == movie_name].index[0]

    movie_vector = tfidf_matrix[idx]

    distances, indices = knn.kneighbors(movie_vector, n_neighbors=11)

    similar_movies = indices[0][1:]

    print(f"\nRecommendations for '{movie_name.title()}':\n")
    for i in similar_movies:
        print(df['title'].iloc[i])

```

```python
recommend("Ava",df,tfidf_matrix_holly,knn_holly)
```

```python
recommend("kabi",df_bolly,tfidf_matrix_bolly,knn_bolly)
```
